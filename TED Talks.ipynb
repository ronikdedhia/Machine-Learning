```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('dataset/ted-talks.csv')
data.head()
```

```python
data.info()
```

```python
data[pd.isnull(data['author'])]
```

```python
data["author"].fillna("Unknown", inplace = True)
```

```python
# no of unique authors
len(data['author'].unique())
```

```python
data.sort_values(by="views", ascending=False)[:10]
```

```python
top_by_views = data.sort_values(by="likes", ascending=False)[:10]
fig, ax = plt.subplots(figsize=(12,6))
ax.bar(top_by_views['title'],top_by_views['views'])
labels = ax.get_xticklabels()
plt.setp(labels, rotation=45, horizontalalignment='right')
plt.xlabel('Authors')
plt.ylabel('Views')
plt.title('Most Popular Video')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12,6))
ax.bar(top_by_views['title'],top_by_views['likes'])
labels = ax.get_xticklabels()
plt.setp(labels, rotation=45, horizontalalignment='right')
plt.title('Most Liked')
plt.xlabel('Authors')
plt.ylabel('Likes')
plt.show()
```

```python
data["Ratio"] = 1/ (data.likes / data.views)
data.sort_values(by="Ratio", ascending=False)[:10]
```

```python
top_by_views = data.sort_values(by="Ratio", ascending=False)[:10]
ax = top_by_views.plot.bar(x ="title", y="Ratio",rot=90, figsize=(12,6))
plt.title("Like/Views Ratio")
plt.ylim([35, 37])
plt.xlabel('Authors')
plt.ylabel('Percentange of Likes/Views')
plt.show()
```

```python
top_author = data.groupby("author").agg("count")["title"].sort_values(ascending=False)[:10]
top_author.plot.bar(stacked=True,rot=45,figsize=(12,6))
plt.title('Top Authors')
plt.xlabel('Authors')
plt.ylabel('No of Shows')
plt.show()
```

```python
data["Month"] = data.date.apply(lambda x: x.split()[0])
data["Year"] = data.date.apply(lambda x: x.split()[1])
data.head()
```

```python
showonmonths = data["Month"].value_counts()
showonmonths.plot.bar(stacked=True,rot=45,figsize=(12,6))
plt.title("Popular Months")
plt.xlabel('Months')
plt.ylabel('Views')
plt.show()
```

```python
showonyears = data["Year"].value_counts().sort_index(ascending=False)
showonyears.plot.bar(stacked=True,rot=45,figsize=(12,6))
plt.title("Popular Year", fontsize=20)
plt.xlabel('Years')
plt.ylabel('Views')
plt.show()
```

```python
from wordcloud import WordCloud, STOPWORDS
stopwords = set(STOPWORDS)
print(stopwords)
```

```python
comment_words = ''
for line in data.title:
    words = line.lower().split()
    comment_words += " ".join(words)+" "
```

```python
wordcloud = WordCloud(width = 1000, height = 800,
                background_color ='black',
                stopwords = stopwords,
                min_font_size = 10).generate(comment_words)
```

```python
plt.figure(figsize = (8, 8), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)
plt.show()
```
