```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings('ignore')
sns.set_style('darkgrid')
```

```python
df=pd.read_csv('dataset/haberman.csv')
df.columns = ['age', 'year', 'nodes', 'status']
df.head()
```

```python
df.info()
```

```python
df['status'].value_counts()
```

```python
status = df.status.value_counts()
plt.figure(figsize=(8, 5))
sns.barplot(status.index, status)
plt.xlabel('Status', fontsize=15);
plt.ylabel('Count', fontsize=15);
```

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(x='age', y='nodes', data=df)
plt.xlabel('age', fontsize=15)
plt.ylabel('nodes', fontsize=15);
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True)
plt.title('Correlation Matrix', fontsize=20);
```

```python
''' Multivariate Analysis '''
plt.figure(figsize=(20, 8))
sns.pairplot(df, hue='status');
```

```python
for i, column in enumerate(df.columns):
    sns.distplot(df[column])
    plt.title(column, fontsize=20);
    plt.show()
```

```python
for column in df.columns:
    plt.figure(figsize=(10, 5))
    sns.boxplot(df[column])
    plt.title(column, fontsize=20);
```

```python
for column in df.columns:
    plt.figure(figsize=(10, 5))
    sns.histplot(df[column], kde=True)
    plt.title('Histogram plot of {}'.format(column), fontsize=20);
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x='status',y='nodes',data=df)
plt.title('Box plot', fontsize=20);
plt.xlabel('status', fontsize=15)
plt.ylabel('nodes', fontsize=15)
plt.show();
```

```python
plt.figure(figsize=(10, 5))
sns.violinplot(x='status',y='nodes',data=df)
plt.title('Violin plot', fontsize=20)
plt.xlabel('status', fontsize=15);
plt.ylabel('nodes', fontsize=15);
```

```python
plt.figure(figsize=(10, 5))
sns.jointplot(x='status', y='nodes', data=df, kind='kde')
plt.title('Univariate Analysis')
plt.xlabel('staus', fontsize=15)
plt.ylabel('nodes', fontsize=15);
```

```python
sns.jointplot(x='age',y='nodes',data=df,kind='kde')
plt.legend
plt.title('Joint plot or Contour plot')
plt.show
```

```python
sns.set_style('whitegrid')
sns.FacetGrid(df,hue='status',height=5) \
.map(plt.scatter,'age','nodes')\
.add_legend()
plt.title('Scatterplot using seasborn of age vs nodes')
plt.show()
```

```python
sns.set_style('whitegrid')
sns.pairplot(df,hue='status',height=3)
plt.show()
```

```python
sns.FacetGrid(df,hue='status', height=8)\
.map(sns.distplot,'age')\
.add_legend()
plt.title('Histogram of age')
```

```python
sns.FacetGrid(df,hue='status',height=8)\
.map(sns.distplot,'nodes')\
.add_legend()
plt.title('Histogram of nodes')
plt.show()
```

```python
sns.FacetGrid(df,hue='status',height=8)\
.map(sns.distplot,'year')\
.add_legend()
plt.title('Histogram of Year')
plt.show()
```

```python
counts, bin_edges = np.histogram(df['nodes'], bins=10, density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges);
cdf = np.cumsum(pdf)
arr1,=plt.plot(bin_edges[1:],pdf);
arr2,=plt.plot(bin_edges[1:],cdf)
plt.title('CDF and PDF of people living more than 5 years')
plt.legend([arr1,arr2], ['PDF survived','CDF survived'])
plt.ylabel('count')
plt.xlabel('nodes')
```
