```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data1 = pd.read_csv('data.csv', encoding= 'unicode_escape')
data1
```

```python
data1['month'] = pd.DatetimeIndex(data1['InvoiceDate']).month
data1 = data1.drop(['InvoiceNo', 'StockCode', 'Description', 'Country'], axis=1)
data1
```

```python
data1.loc[data1.month == 3]
```

```python
data1.Quantity.describe()
```

```python
# there are values in quantity that are less than 0
# thus we remove the columns having those values
data1 = data1[data1['Quantity']>0]
data1.Quantity.describe()
```

```python
data1 = data1[data1['UnitPrice']>0]
data1.UnitPrice.describe()
```

```python
# unit price seems to be regular and normal
```

```python
data1.isnull().sum()
```

```python
# there are null values present in customerId in the dataset
# since there is no way to impute those values, we choose to drop them
data1.dropna(inplace=True)
data1.isnull().sum()
```

```python
data1['PurchaseAmount'] = data1['UnitPrice'] * data1['Quantity']
data1
```

```python
data1 = data1.drop(['Quantity', 'InvoiceDate', 'UnitPrice'], axis=1)
data1
```

```python
df_agg = data1.groupby(['month', 'CustomerID']).agg({'PurchaseAmount': np.sum}).reset_index()
df_agg.head()
```

```python
from sklearn.model_selection import train_test_split
X = df_agg.drop(['PurchaseAmount'], axis=1)
y = df_agg['PurchaseAmount']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
y.head()
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X_train, y_train)
predictions = lm.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
print(mean_squared_error(y_test, predictions))
print(lm.score(X_test, y_test))
rmse = (np.sqrt(mean_squared_error(y_test, predictions)))
print("RMSE is {}", format(rmse))
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators=10, random_state=0)
regressor.fit(X_train, y_train)
```

```python
predictions = regressor.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
print(mean_squared_error(y_test, predictions))
print(lm.score(X_test, y_test))
rmse = (np.sqrt(mean_squared_error(y_test, predictions)))
print("RMSE is {}", format(rmse))
```
