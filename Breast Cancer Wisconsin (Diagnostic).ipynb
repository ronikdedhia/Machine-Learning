```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv('dataset/breast_cancer.csv')
data.head()
```

```python
data.columns
```

```python
data.drop(['Unnamed: 32', 'id'], axis = 1, inplace = True)
```

```python
data.describe()
```

```python
data['diagnosis'].unique()
```

```python
fig, ax = plt.subplots(figsize=(20, 10))
sns.heatmap(data.corr(), cmap='RdBu', annot=True, fmt=".2f");
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
x = data.drop('diagnosis',axis=1)
y = data['diagnosis']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
rf = RandomForestClassifier()
rf = rf.fit(x_train,y_train)
predictions = rf.predict(x_test)
```

```python
print(accuracy_score(y_test, predictions)*100)
```

```python
# dropping the columns that have least effect on the output and removing them will make a lighter model
drop_list1 = ['perimeter_mean','radius_mean','compactness_mean','concave points_mean','radius_se','perimeter_se','radius_worst','perimeter_worst','compactness_worst','concave points_worst','compactness_se','concave points_se','texture_worst','area_worst']
x= x.drop(drop_list1,axis=1)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
rf = RandomForestClassifier()
rf = rf.fit(x_train,y_train)
predictions = rf.predict(x_test)
```

```python
print(accuracy_score(y_test, predictions)*100)
```
